```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
x1=[10,11,23,34,14]
plt.stackplot(x,x1)
plt.show()
```

```python
x=[2,3,4,5,6]
area1=[2,4,6,3,1]
area2=[3,4,2,5,6]
area3=[5,4,1,2,3]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,colors=["r","k","m"],labels=l)
plt.legend(loc=1)  #if set loc=2 then labels turn into the left side
plt.show()
```

```python
# baseline parameter
x=[2,3,4,5,6]
area1=[2,4,6,3,1]
area2=[3,4,2,5,6]
area3=[5,4,1,2,3]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,colors=["r","k","m"],labels=l,baseline="sym")
plt.legend(loc=1)  #if set loc=2 then labels turn into the left side
plt.show()
```

```python
x=[2,3,4,5,6]
area1=[2,4,6,3,1]
area2=[3,4,2,5,6]
area3=[5,4,1,2,3]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,colors=["r","k","m"],labels=l)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=1)  #if set loc=2 then labels turn into the left side
plt.show()
```
