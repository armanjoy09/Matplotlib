```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,4,5,3,7,8]
plt.plot(x,y,color="r")
plt.xticks(x)
plt.yticks(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,4,5,3,7,8]
plt.plot(x,y,color="r")
plt.xticks(x,labels=["python","c++","c","java","R","latex"])
plt.yticks(y)
plt.show()
```
