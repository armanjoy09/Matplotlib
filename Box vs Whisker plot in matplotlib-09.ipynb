```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x)
plt.show()
```

```python
# for changing the shape of the box
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,notch=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,notch=False)
plt.show()
```

```python
# to change the vartical and horizontal move of  the box use vert=True/False


x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,notch=True,vert=False)
plt.show()

```

```python
# change the width of the box
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,widths=0.8)  #default value is 0.2
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,notch=True,widths=0.5,vert=False)
plt.show()
```

```python
# labels of the boxplot
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,labels=["python"])
plt.show()
```

```python
# to fill the  box use patch_artist=True/False
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,labels=["python"],patch_artist=True)
plt.show()
```

```python
# means of the box plot
x=[10,20,30,40,50,60,70,80,90]
plt.boxplot(x,labels=["python"],patch_artist=True,showmeans=True)
plt.show()
```

```python
# outlier of the box
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["python"])
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120,130,140]
plt.boxplot(x,labels=["python"])
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,150,160]
plt.boxplot(x,labels=["python"])
plt.show()
```

```python
# lenth of the whisker plot
x=[10,20,30,40,50,60,70,80,150,160]
plt.boxplot(x,labels=["python"],whis=2.5)
plt.show()
```

```python
# color of the whisker plot
x=[10,20,30,40,50,60,70,80,150,200]
plt.boxplot(x,labels=["python"],sym="g+")
plt.show()
```

```python
# if want to hide the whisker plot then
x=[10,20,30,40,50,60,70,80,200]
plt.boxplot(x,labels=["python"],sym=" ")
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,150,200,250]
plt.boxplot(x,labels=["python"],sym="m+")
plt.show()
```

```python
# to change the properties of the box ,cap,whisker,filler
x=[10,20,30,40,50,60,70,80]
plt.boxplot(x,showmeans=True,boxprops={"color":"r"},capprops={"color":"m"},
            whiskerprops={"color":"b"},
            flierprops={"markeredgecolor":"k"} )
plt.show()
```

```python
# create  multiple boxplot in a plot
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,80]
x1=[30,40,50,60,70,80,90,100,110,120]
plt.boxplot(x,showmeans=True,boxprops={"color":"r"},capprops={"color":"m"},
            whiskerprops={"color":"b"})
plt.boxplot(x1,showmeans=True,boxprops={"color":"r"},capprops={"color":"m"},
            whiskerprops={"color":"b"})
plt.show()     #thats get overlap one into one
```

```python
# we create multiple box plot in this way
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,80]
x1=[30,40,50,60,70,80,90,100,110,120]
y=[x,x1]
plt.boxplot(y,showmeans=True,labels=["python","c++"],boxprops={"color":"r"},capprops={"color":"m"},
            whiskerprops={"color":"b"})
plt.show()
```
