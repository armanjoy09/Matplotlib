```python
# basic fill_between plot
import matplotlib.pyplot as plt

x=[3,4,5,6,7,8]
y=[3,4,5,6,7,8]

plt.plot(x,y,color="r")
plt.fill_between(x,y,color="m",alpha=0.5)
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# if want to fill specific region of the plot then 
import matplotlib.pyplot as plt
x=[2,3,4,5,6,7]
y=[2,3,4,5,6,7]
plt.plot(x,y,color="r")
plt.fill_between(x=[4,6],y1=4,y2=6)  #in this we have to put the value of y1 and y2 according to x
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# if i want to make a specific region fill without putting the positional argument
# then we need to put where argument that helps to specify the fixed regiion of the plot

import matplotlib.pyplot as plt
import numpy as np
x=np.array([3,4,5,6,7,8])
y=np.array([3,4,5,6,7,8])
plt.plot(x,y)
plt.fill_between(x,y,where=(x>=4)&(x<=6))  #('>=' not supported between instances 
plt.title("Python")                      #of 'list' and 'int')for this reseaon we convert the list
                                         #into array using numpy
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([3,4,5,6,7,8])
y=np.array([3,4,5,6,7,8])
plt.plot(x,y)
plt.fill_between(x,y,where=(x>=4)&(x<=6),alpha=0.4,color="b") 
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
