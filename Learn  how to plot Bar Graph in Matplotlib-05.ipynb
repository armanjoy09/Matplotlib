```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[85,70,75,80]
c=['r','b','m','k']
plt.xlabel('language',fontsize=15)
plt.ylabel('marks',fontsize=20)
plt.title('object oriented language',fontsize=23)
plt.bar(x,y,color=c,width=0.4,align='center',edgecolor='y',linewidth=5,linestyle=":",alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['python','c','c++','java']
y=[85,70,75,80]
z=[40,49,35,58]
t=[50,44,76,86]
width=0.2
var=np.arange(len(x))
var1=[i+width for i in var]
var2=[i+width for i in var1]

c=['r','b','m','k']
plt.xlabel('language',fontsize=15)
plt.ylabel('marks',fontsize=20)
plt.title('object oriented language',fontsize=23)
plt.bar(var,y,width,color=c,label="popularity")
plt.bar(var1,z,width,color='y',label="popularity1")
plt.bar(var2,t,width,color='g',label="popularity2")
plt.xticks(var+width/2,x)
plt.legend()
plt.show()
```
