# Bar Plot

```python
# basic bar plot
import matplotlib.pyplot as plt
x=['python','c','c++','Java']
y=[85,70,75,80]
plt.bar(x,y)
plt.show()
```

```python
# if want to differnet color bar plot then
x=['python','c','c++','java']
y=[85,70,75,80]
c=['m','g','y','r']
plt.bar(x,y,color=c)
plt.show()
```

```python
# if i want to labeling the axis and tittle then
x=['python','c','c++','java']
y=[85,70,75,80]
plt.xlabel("language")
plt.ylabel("marks")
plt.title("object oriented laguages")
plt.bar(x,y,color=c)
plt.show()
```

```python
# if i want to change the front of label and title name then
x=['python','c','c++','java']
y=[85,70,75,80]
c=['m','g','y','r']
plt.xlabel("language",fontsize=15)
plt.ylabel("marks",fontsize=12)
plt.title("object oriented language",fontsize=20)
plt.bar(x,y,color=c)
plt.show()
```

```python
# if i want to change the width of the bar then
x=['python','c','c++','java']
y=[85,70,75,80]
c=['m','g','y','r']
plt.xlabel("language",fontsize=15)
plt.ylabel("marks",fontsize=12)
plt.title("object oriented language",fontsize=20)
plt.bar(x,y,width=0.5,color=c)
plt.show()
```

```python
# if i want to change the language indicatore  into the one of adge from center then
# it remains in the center by default
x=['python','c','c++','java']
y=[85,70,75,80]
c=['m','g','y','r']
plt.xlabel("language",fontsize=15)
plt.ylabel("marks",fontsize=12)
plt.title("object oriented language",fontsize=20)
plt.bar(x,y,color=c,align="edge",width=0.5)
plt.show()
```

```python
# if want to change the bar boundary line color or edge color then
x=['python','c','c++','java']
y=[85,70,75,80]
c=['m','g','y','r']
plt.xlabel("language",fontsize=15)
plt.ylabel("marks",fontsize=12)
plt.title("object oriented language",fontsize=20)
plt.bar(x,y,color=c,width=0.5,edgecolor='k',)
plt.show()
```

```python
# if want to change the width of edge then
x=['python','c','c++','java']
y=[85,70,75,80]
c=['m','g','y','r']
plt.xlabel("language",fontsize=15)
plt.ylabel("marks",fontsize=12)
plt.title("object oriented language",fontsize=20)
plt.bar(x,y,color=c,width=0.5,edgecolor='k',linewidth=5)
plt.show()
```

```python
# if want to change the line style of the edge(there are many line style) into dot dot then
x=['python','c','c++','java']
y=[85,70,75,80]
c=['m','g','y','r']
plt.xlabel("language",fontsize=15)
plt.ylabel("marks",fontsize=12)
plt.title("object oriented language",fontsize=20)
plt.bar(x,y,color=c,width=0.5,edgecolor='k',linewidth=5,linestyle=":")
plt.show()
```

```python
# if want to change the bar color into light color  then use alpha parameter
x=['python','c','c++','java']
y=[85,70,75,80]
c=['m','g','y','r']
plt.xlabel("language",fontsize=15)
plt.ylabel("marks",fontsize=12)
plt.title("object oriented language",fontsize=20)
plt.bar(x,y,color=c,width=0.5,edgecolor='k',linewidth=5,linestyle=":",alpha=0.5)
plt.show()
```

```python
# if want to display label in the bar plot
x=['python','c','c++','java']
y=[85,70,75,80]
c=['m','g','y','r']
plt.xlabel("language",fontsize=15)
plt.ylabel("marks",fontsize=12)
plt.title("object oriented language",fontsize=20)
plt.bar(x,y,color='m',width=0.5,label="popularity")
plt.legend()
plt.show()
```

```python
# multiple bar chart

import matplotlib.pyplot as plt
import numpy as np

x=['python','c','c++','java']
y=[85,70,75,80]
ML=[90,50,60,80]
web=[82,95,98,92]

width=0.2
p=np.arange(len(x))
p1=[i+width for i in p]
p2=[i+width for i in p1]

plt.xlabel("language")
plt.ylabel("marks")
plt.title("object oriented language")

plt.bar(p,y,width,color="b",label="popularity")
plt.bar(p1,ML,width,color='m',label="popularity1")
plt.bar(p2,web,width,color='r',label="popularity2")

plt.xticks(p+width,x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=['python','c','c++','java']
y=[85,70,75,80]
ML=[90,50,60,80]

width=0.2
p=np.arange(len(x))
p1=[i+width for i in p]
plt.xlabel("language")
plt.ylabel("marks")
plt.title("object oriented language")

plt.bar(p,y,width,color="b",label="popularity")
plt.bar(p1,ML,width,color='m',label="popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
countries=['USA','RUSSIA','CANADA','GERMANEY','AUSTRALIA']
Gold=[50,30,49,59,70]
Silver=[40,25,58,68,53]
Bronze=[34,67,23,45,90]

width=0.2
v=np.arange(len(countries))
v2=[i+width for i in v]
v3=[i+width for i in v2]

plt.xlabel("Country",fontsize=15)
plt.ylabel("Medal",fontsize=16)
plt.title("Rio 2016 oylimpic games",fontsize=20)

plt.bar(v,Gold,width,color='m',label="Gold")
plt.bar(v2,Silver,width,color='r',label="Silver")
plt.bar(v3,Bronze,width,color='b',label="Bronze")

plt.xticks(v+width,countries)
plt.legend()
plt.show()
```
