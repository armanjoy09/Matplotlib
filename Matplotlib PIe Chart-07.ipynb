```python
import matplotlib.pyplot as plt
x=['python','java','c++','c']
y=[40,30,20,10]
c=['r','m','b','g']
plt.pie(y,labels=x,colors=c)
plt.show()
```

```python
# if we use explode function
import matplotlib.pyplot as plt
x=[40,30,20,10]
y=['python','java','c++','c']
c=['m','r','y','g']
ex=[0.0,0.0,0.0,0.2]
plt.pie(x,labels=y,colors=c,explode=ex)
plt.show()
```

```python
# show the parcentage of these graph use the function name, autopct="%(value)f%%"

import matplotlib.pyplot as plt
x=[20,30,40,50]
y=['c','c++','java','python']
c=['m','g','y','r']
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,colors=c,labels=y,explode=ex,autopct="%0.1f%%")
plt.show()
```

```python
# if want to see the shadow of the graph then use shadow function
x=[20,30,40,50]
y=['c','c++','java','python']
c=['m','g','y','r']
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,colors=c,labels=y,explode=ex,autopct="%0.2f%%",shadow=True)  #here we can see the shadow of the graph
plt.show()
```

```python
# if want to increase or decrease the pie circle then use the radius function
x=[20,30,40,50]
y=['c','c++','java','python']
c=['m','g','y','r']
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,colors=c,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,radius=1.5)
plt.show()
```

```python
# startangle
# startangle is the parameter use to start the pie chart from desire angle or kon
x=[20,30,40,50]
y=['c','c++','java','python']
c=['m','g','y','r']
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,colors=c,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,labeldistance=0.3,startangle=180)
plt.show()
```

```python
# font size of the labels
# for changing the font size of the label then make dictionary in the textprops
x=[20,30,40,50]
y=['c','c++','java','python']
c=['m','g','y','r']
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,colors=c,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,textprops={"fontsize":25})
plt.show()
```

```python
# counterclock

x=[20,30,40,50]
y=['c','c++','java','python']
c=['m','g','y','r']
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,colors=c,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,counterclock=False)
plt.show()
```

```python
# wedgeprops
x=[20,30,40,50]
y=['c','c++','java','python']
c=['m','g','y','r']
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,colors=c,labels=y,explode=ex,autopct="%0.1f%%",shadow=True
       ,wedgeprops={"linewidth":2,"edgecolor":'k',"linestyle":":"})
plt.show()
```

```python
# give title and font size
x=[20,30,40,50]
y=['c','c++','java','python']
c=['m','g','y','r']
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,colors=c,labels=y,radius=1,explode=ex,autopct="%0.1f%%")
plt.title("language popularity",fontsize=20)
plt.show()
```

```python
# legend() funnction
x=[20,30,40,50]
y=['c','c++','java','python']
c=['m','g','y','r']
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,colors=c,labels=y,radius=1,explode=ex,autopct="%0.1f%%")
plt.title("language popularity",fontsize=20)
plt.legend(loc=2)
plt.show()
```

```python
# dot pie chart
plt.pie([1],radius=0.7)
plt.show()
```

```python
# donat pie chart means multiple pie chart(pie chart into the pie chart)
import matplotlib.pyplot as plt
x=[40,30,20,10]
z=[10,20,30,40]
y=['python','java','c++','c']
c=['m','r','k','w']

plt.pie(x,labels=y,radius=1.5)
plt.pie(z,radius=1,colors=c)
plt.show()
```

```python
# dot pie chart in the pie chart
x=[10,20,30,40]
y=['python','java','c++','c']
c=['y','g','r','m']
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors='w')
plt.show()
```
