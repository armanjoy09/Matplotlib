```python
# basic step plot
import matplotlib.pyplot as plt
x=[2,3,4,5,6,7,8]
y=[3,5,4,6,3,2,7]

plt.step(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[4,5,6,7,8]

plt.step(x,y)
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

```python
# if i want to set the marker then
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[4,5,3,7,8]

plt.step(x,y,color="r",marker="o",ms="10",mfc="k",label="python")
  #color=step line color,marker=gulla,ms=marker size,mfc=marker face color

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[3,4,5,6,7,8]
y=[5,4,3,7,6,1]
plt.step(x,y,marker="d",ms="10",mfc="r",color="k",label="Python")
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2)
plt.grid()
plt.show()
```
