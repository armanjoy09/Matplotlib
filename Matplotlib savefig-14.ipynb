```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[2,3,4,5,6]
plt.plot(x,y,color="r")
plt.savefig("linear linebasic")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[2,3,4,5,6]
plt.plot(x,y,color="r")
plt.savefig("linear line3",dpi=1000)  #dpi=dot per inches that increase zoom,clearity of figure
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[2,3,4,5,6]
plt.plot(x,y,color="r")
plt.savefig("linear line4",dpi=2000,facecolor="m")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[2,3,4,5,6]
plt.plot(x,y,color="r")
plt.savefig("linear line5",dpi=2000,facecolor="g",transparent=True)
plt.show()
```

```python
# use bbox_inches 

import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[2,3,4,5,6]
plt.plot(x,y,color="r")
plt.savefig("linear line6",dpi=2000,facecolor="b",transparent=False
            ,bbox_inches="tight")
plt.show()
```
