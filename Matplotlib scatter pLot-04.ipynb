```python
# basic scatter plot
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[9,4,5,6,3]
plt.scatter(x,y)
plt.show()
```

```python
# if i want to change the color of each plot then
x=[2,3,4,5,6]
y=[4,5,3,1,5]
c=['k','r','g','y','m']
plt.scatter(x,y,color=c)
plt.show()
```

```python
# if i want to change the size of each plot
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
c=['m','r','g','k','b','r']
sizes=[100,200,300,250,400,350]
plt.scatter(day,No,s=sizes,color=c)
plt.show()
```

```python
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
c=['m','r','g','k','b','r']
sizes=[50,20,100,200,150,300]
plt.scatter(day,No,s=sizes,color=c)
plt.show()
```

```python
# if i want to add title and label of x ,y then their fontsize
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
c=['m','r','g','k','b','r']
sizes=[100,200,300,250,400,350]
plt.scatter(day,No,s=sizes,color=c)
plt.title("scatter plot",fontsize=25)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=18)
plt.show()
```

```python
# if i want to blur the scatter then use alpha function
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
c=['m','r','g','k','b','r']
sizes=[100,200,300,250,400,350]
plt.scatter(day,No,s=sizes,color=c,alpha=0.5)
plt.title("scatter plot",fontsize=25)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=18)
plt.show()
```

```python
# if i want to change the edgecolor then
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
c=['m','r','g','k','b','r']
sizes=[100,200,300,250,400,350]
plt.scatter(day,No,s=sizes,color=c,edgecolor='k')
plt.title("scatter plot",fontsize=25)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=18)
plt.show()
```

```python
# if i want to change the edge width then
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
c=['m','r','g','k','b','r']
sizes=[100,200,300,250,400,350]
plt.scatter(day,No,s=sizes,color=c,edgecolor='k',linewidth=3)
plt.title("scatter plot",fontsize=25)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=18)
plt.show()
```

```python
# if i want to change the line style of edge then
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
c=['m','r','g','k','b','r']
sizes=[100,200,300,250,400,350]
plt.scatter(day,No,s=sizes,color=c,edgecolor='k',linewidth=3,linestyle=":")
plt.title("scatter plot",fontsize=25)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=18)
plt.show()
```

```python
# if i want to change the marker of scatter plot then use marker parameter

# if i want to change the edgecolor then
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
c=['m','r','g','k','b','r']
sizes=[250,200,300,250,400,350]
plt.scatter(day,No,s=sizes,color=c,marker="*")
plt.title("scatter plot",fontsize=25)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=18)
plt.show()
```

```python
# gives the different color for scatter plot by color map(cmap function)
# if i want to change the edgecolor then
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
color=[10,20,30,40,45,50]
sizes=[100,200,300,250,400,350]
plt.scatter(day,No,s=sizes,c=color,cmap="viridis",)
plt.colorbar()
plt.title("scatter plot",fontsize=25)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=18)
plt.show()
```

```python
# if i want to labeling the color bar then 
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
color=[10,20,30,40,25,45]
sizes=[100,200,300,250,400,350]
plt.scatter(day,No,s=sizes,c=color,alpha=0.5)
t=plt.colorbar()
t.set_label("colorbar",fontsize=20)
plt.title("scatter plot",fontsize=25)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=18)
plt.show()
```

```python
# make a multiple scatter plot in single plot
# if i want to change the edgecolor then
day=[2,4,5,6,7,8]
No=[1,3,5,6,7,9]
No2=[3,6,7,4,9,1]
c=['m','r','g','k','b','r']
sizes=[100,200,300,250,400,350]
plt.scatter(day,No,s=sizes,color=c)
plt.scatter(day,No2,s=sizes,color='r')
plt.title("scatter plot",fontsize=25)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=18)
plt.show()
```
