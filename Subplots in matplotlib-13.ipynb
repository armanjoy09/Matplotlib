```python
import matplotlib.pyplot as plt

x=[2,3,4,5,6]
y=[2,3,4,5,6]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")       #here (2,2,1) is (nrows,ncols,index)


plt.subplot(2,2,2)
plt.pie([1])


x1=[10,20,30,40]
plt.subplot(2,2,3)        #subplot  main plot ba pie plot ba bar plot er  age dite hobe
plt.pie(x1)



x2=['arman','Joy','nehal','arhan']
x3=[70,80,90,84]
plt.subplot(2,2,4)
plt.bar(x2,x3) 


plt.show()
```
