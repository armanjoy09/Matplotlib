```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[3,5,2,6,3]
plt.title("Python",fontsize=15)
plt.xlabel("Days",fontsize=12)
plt.ylabel("language",fontsize=12)
plt.plot(x,y)
plt.text(4,5,"Java",fontsize=15,style="italic",bbox={"facecolor":"red"})   #here 4,5 is the value
                                                                          #of x and y, 'java'word
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[3,5,2,6,3]
plt.title("Python",fontsize=15)
plt.xlabel("Days",fontsize=12)
plt.ylabel("language",fontsize=12)
plt.plot(x,y)
plt.text(4,5,"Java",fontsize=15,style="italic",bbox={"facecolor":"red"})
plt.annotate("python",xy=(4,2),xytext=(5,4),arrowprops={"facecolor":"black","shrink":100})
plt.show()
```
